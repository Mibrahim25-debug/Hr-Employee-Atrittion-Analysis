```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("C:/Users/hp/Downloads/Hr-employee.csv")

#checking for missing values now
df.isnull().sum()
df.columns = df.columns.str.strip()

#Now checking for data types
df.info()
df.dtypes

#checking for duplicated rows
df.duplicated().sum()

#checking columns renaming if needed and converting their types
df.columns = df.columns.str.strip().str.replace(" ", "_").str.lower()
df.drop(columns=['employeecount', 'over18', 'standardhours'], inplace=True)
cat_columns = ["attrition", "businesstravel", "department", "educationfield",
    "gender", "jobrole", "maritalstatus", "overtime"]

for col in cat_columns:
    df[col] = df[col].astype("category")

df.info()

#Performing EDA 
#How many number of employees left or stayed according to each gender group
attrition_by_gender = df.groupby("gender")["attrition"].value_counts(normalize = True).unstack() * 100
attrition_by_department = df.groupby("department")["attrition"].value_counts(normalize = True).unstack() * 100
attrition_by_jobrole = df.groupby("jobrole")["attrition"].value_counts(normalize = True).unstack() * 100
print(attrition_by_gender)
print(attrition_by_department)
print(attrition_by_jobrole)
#Now finding the correlation between salary, satisfaction and attrition
corr_df = df[['monthlyincome', 'jobsatisfaction', 'attrition']].copy()
corr_df["attrition"] = corr_df["attrition"].map({'Yes' : 1, 'No' : 0})
corr_matrix = corr_df.corr()
print(corr_matrix)

#Performing visulizations now
#Countplot for categorical data
plt.figure(figsize = (6,4))
sns.countplot(x = "gender", hue = "attrition", data = df)
plt.title("Attrition count by gender")
plt.show()

#Box plot 
plt.figure(figsize = (6,4))
sns.boxplot(x = "attrition", y = "monthlyincome", data = df)
plt.title("Monthly income VS Attrition")
plt.show()

#NOW the heatmap
selected_cols = ['age', 'monthlyincome', 'joblevel', 'jobsatisfaction',
                 'environmentsatisfaction', 'performancerating', 'worklifebalance',
                 'yearsatcompany', 'totalworkingyears']

plt.figure(figsize=(8,6))
sns.heatmap(df[selected_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Focused Correlation Heatmap (HR Insights)')
plt.show()

#now exporting the CSV file for further analysis using sql
df.to_csv("cleaned_hr_data.csv", index = False)
```
